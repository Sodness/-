```python
#워드 클라우드
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'
# 유니코드에서  음수 부호설정
plt.rc('axes', unicode_minus=False)
```

```python
df_raw = pd.read_excel('./빅카인즈_태양광.xlsx')
df = df_raw.copy()
```

```python
stopwords = ['재생 에너지', '신재생 에너지', '재생', '에너지', '신재생']
```

```python
res = []
for i in df['키워드']:
    # print(i)
    res.append(i)
    
tmp = ','.join(res)
txt = tmp.replace(',', ' ')
```

```python
wc = WordCloud('./NanumGothic.ttf',
               mode='RGBA',
               background_color = None,
               max_words = 200,
               stopwords = stopwords)
wc = wc.generate(txt) # str자료만 가능, list자료는 안됨
plt.figure(figsize = (12, 12))
plt.imshow(wc) #interpolation = "bilinear")
plt.axis("off")

# plt.savefig('빅카인즈_태양광_워드클라우드.png')
plt.show()
```
